```python
def extract_adjectives_total(model_input):
    
    counter = 0
    
    time.sleep(3.1)
    
    if type(model_input) == str:
    
        if len(model_input.split())>4000:
            #Counts the number of requests depending on the token length of the string
            nr_of_requests = len(model_input.split())/1000
            counter+=nr_of_requests

            #Splits the model_inputs in evenly big chunks that are each smaller than 4,000
            chunks = chunk_text(model_input, len(model_input.split())/math.ceil(len(model_input.split())/4000))
            summaries = extract_adjectives(chunks)
            text = "\n\n".join(summaries)
            return text

        elif len(model_input.split())<=4000:
            #Counts the number of requests depending on the token length of the string
            nr_of_requests = len(model_input.split())/1000
            counter+=nr_of_requests

            summaries = extract_adjectives(model_input)
            return summaries

        else:
            print("Error in extract_adjectives_total")
    
    else:
        pass
```

```python
def extract_adjectives(chunks):
    
    adjectives = []
    
    #If the number of reviews is less than 4000 tokens, the type should be a string and hence only one request is sent to the Open AI API
    if type(chunks) == str:
        try:
            completion = openai.ChatCompletion.create(
                  model="gpt-3.5-turbo", 
                  messages=[{"role": "user", "content": f"Imagine you are a restaurant critic. Please summarise the following restaurant reviews and information into the best descriptive adjectives. Respect the following guidelines: Ignore all occurrences of '\'. Each review has the following format: ‘review0’ : ‘text_of_review0‘, … , ‘review4’ : ‘text_of_review4‘, ‘editorial_summary’ : ‘text_of_summary‘, ‘types’ : [ ‘type_0’, …, ‘type_last’]. Order the adjectives by importance as determined by the topic overlap of all the reviews plus the editorial_summary and the types. Explicitly assess if the adjective “hidden” is present in the reviews The overall output format should be: 'descriptive adjectives: list of adjectives, hidden: No/Yes'\n Reviews: {chunks}"}]
                    )

            adjectives.append(completion["choices"][0]["message"]["content"])
        except:
            adjectives.append("")
    
    
    #If the number of reviews is more than 4000 tokens, the type should be a list (model input is got chunked before in more parts) and hence more than one request is sent to the Open AI API
    elif type(chunks) != str:
        for chunk in chunks:
            try:
                completion = openai.ChatCompletion.create(
                  model="gpt-3.5-turbo", 
                  messages=[{"role": "user", "content": f"Imagine you are a restaurant critic. Please summarise the following restaurant reviews and information into the best descriptive adjectives. Respect the following guidelines: Ignore all occurrences of '\'. Each review has the following format: ‘review0’ : ‘text_of_review0‘, … , ‘review4’ : ‘text_of_review4‘, ‘editorial_summary’ : ‘text_of_summary‘, ‘types’ : [ ‘type_0’, …, ‘type_last’]. Order the adjectives by importance as determined by the topic overlap of all the reviews plus the editorial_summary and the types. Explicitly assess if the adjective “hidden” is present in the reviews The overall output format should be: 'descriptive adjectives: list of adjectives, hidden: No/Yes'\n Reviews: {chunk}"}]
                    )

                adjectives.append(completion["choices"][0]["message"]["content"])
            except:
                adjectives.append("")
                
    else:
        print("Error in extract_adjectives")

    return adjectives
```

```python
#Splits the text in the specified review into chunks of the specified length.
def chunk_text(review, chunk_length):
    try:
        chunks = [review[i:i + chunk_length] for i in range(0, len(review), chunk_length)]
        return chunks
    except:
        pass
```

```python
def determine_type(text):
    return type(text)
```

```python
def uniform_chatgpt_outcome(text):
    if type(text) != list:
        return np.nan
    else:
        return text
```

```python
def transform_empty_list(text):
    if type(text) == list:
        if len(text[0])==0:
            return np.nan
        else:
            return text
    else:
        return text
```

```python
restaurant_list["adjectives"] = restaurant_list["adjectives"].apply(lambda x: transform_empty_list(x))
```

```python
def lower_column(adjective_list):
    if type(adjective_list) == list:
        return [adjective.lower() for adjective in adjective_list]
    else:
        pass
```

```python
restaurant_list["adjectives"] = restaurant_list["adjectives"].apply(lambda x: lower_column(x))
```

```python
def extract_adjectives2(text):
    
    #if text != None or len(text) > 0:
    
    if type(text) == list:
        # find the start and end positions of the adjectives
        start = text[0].find('descriptive adjectives:') + len('descriptive adjectives:')
        end = text[0].find('hidden:')

        # extract the adjectives
        adjectives = [adj.strip() for adj in text[0][start:end].split(',')]

        return adjectives
    
    else:
        return np.nan
```

```python
def extract_hidden_feature(text):
    
    if type(text) == list:
        # find the position of the "hidden" keyword
        hidden_pos = text[0].find('hidden:')

        # extract the hidden feature value
        hidden_value = None
        if hidden_pos != -1:
            hidden_value = text[0][hidden_pos:].split()[-1].strip('.').lower()
            if hidden_value not in ['yes', 'no']:
                hidden_value = None

        return hidden_value
    
    else:
        return np.nan
```

```python
#Pulling descriptive adjectives of ChatGPT prompts
restaurant_list["descriptive_adjectives"]  = restaurant_list["adjectives"].apply(lambda x: extract_adjectives2(x))
```

```python
#Pulling descriptive adjectives of ChatGPT prompts
restaurant_list["hidden?"]  = restaurant_list["adjectives"].apply(lambda x: extract_hidden_feature(x))
```
